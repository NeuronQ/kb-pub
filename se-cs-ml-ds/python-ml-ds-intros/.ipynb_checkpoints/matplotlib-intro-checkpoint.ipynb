## Imports/environment

```python
from IPython.display import display

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

## Some test data

```python
# some data to play with
x = np.arange(0, 10, 0.01)
y1 = np.exp(-x/4) * np.cos(x**2)
y2 = np.sin(x ** 2)
y3 = np.exp(x/4) * np.sin(x**2)
```

## Setting pyplot defaults

```python
plt.rcParams["figure.figsize"][0] = 10
plt.rcParams["figure.figsize"][1] = 4
# to look decent on modern displays (increase to 300 for print)
plt.rcParams["figure.dpi"] = 144
# to avoid bg being transparent and title/axis texts not bieng visible on darg bg.:
plt.rcParams["figure.facecolor"] = "white"
```

## Matlab-stye/implerative API

### How to use Matlab-stye/implerative API

```python
# (optionally) create a figure to set things like size & dpi
plt.figure(figsize=(10, 4), dpi=144)

# title
plt.title('Plotting sutff...', style='italic')

# actual plotting calls: data & labels
plt.plot(x, y1, x, y2, label='demo')
plt.plot(x, y3, label='oscilation OY42', color='g')

# annotations
plt.text(1, 8, r'$e^{\frac{x}{4}}\sin(x^2)$', fontsize=20, color='g')

# vertical and horizontal lines
plt.axvline(1, color='b', linestyle='-.', linewidth=1)
plt.axvline(5, color='tomato', linewidth=2)
plt.axhline(5, color='m', linewidth=1)

# legend
plt.legend(loc='lower left')

# axis label
plt.ylabel('some numbers')

# axis ticks
plt.xticks(np.arange(0,10,2))

# grid
plt.grid(linestyle='-', alpha=0.4)

plt.show()  # not really needed if there's one plot per cell
```

### Why to use Matlab-stye/implerative API
Easier/faster to hack smth. together (even if some more advanced features are only available with the OOP API)

## Modern object-oriented API

### How to use OOP API
1. use `plt.figure` & `plt.add_subplot` (or simpler: `plt.subplots`) to get `Figure` and `Axes`
2. call the methods on `Figure` and `Axes` to get everything else done

```python
# OOP API

# (optionally) create a figure to set things like size & dpi
# - long form:
# fig = plt.figure(figsize=(10, 4), dpi=144)
# ax = fig.add_subplot(1, 1, 1) # same: .add_subplot(111)
# - condensed form:
fig, ax = plt.subplots(1, 1, figsize=(10, 4), dpi=144)

# title
ax.set_title('Plotting sutff...', style='italic')

# actual plotting calls: data & labels
ax.plot(x, y1, x, y2, label='demo')
ax.plot(x, y3, label='oscilation OY42', color='g')

# annotations
ax.text(1, 8, r'$e^{\frac{x}{4}}\sin(x^2)$', fontsize=20, color='g')

# vertical and horizontal lines
ax.axvline(1, color='b', linestyle='-.', linewidth=1)
ax.axvline(5, color='tomato', linewidth=2)
ax.axhline(5, color='m', linewidth=1)

# legend
ax.legend(loc='lower left')

# axis label
ax.set_ylabel('some numbers')

# axis ticks
ax.set_xticks(np.arange(0,10,2))
# example of more advanced feature only available in OOP API:
ax.set_xticks(np.arange(0,10,0.5), minor=True)

# grid
ax.grid(which='major', linestyle='-', alpha=0.4)
ax.grid(which='minor', linestyle=':', color='g')

plt.show()  # not really needed if there's one plot per cell
```

### Why use OOP API

- supports all advanced features
- more maintainable, readable and extensible code
- easy to write helpers for plotting parts of figure

```python
# not the best example, but it's easy to imagine how useful this can be:

def plot_small_oscillations(ax, xs, ys1, ys2):
    ax.plot(xs, ys1, xs, ys2, label='small oscilation')
    ax.axhline(5, color='m', linewidth=1)


def plot_large_oscilation(ax, xs, ys):
    ax.plot(x, y3, label='oscilation OY42', color='g')

fig, ax = plt.subplots(1, 1, figsize=(10, 4), dpi=144)

plot_small_oscillations(ax, x, y1, y2)
plot_large_oscilation(ax, x, y3)
```

## Multiple-plots figures

```python
plt.figure(figsize=(10, 4), dpi=144)
plt.suptitle("Two plots, one figure")

plt.subplot(1, 2, 1)  # 1-row, 2-columns, select 1st
plt.plot(x, y1, x, y2)
plt.title("Foo")

plt.subplot(1, 2, 2)  # 1-row, 2-columns, select 2nd
plt.plot(x, y3)
plt.title("Bar")

plt.tight_layout(pad=2.0)
```

```python
fig, [(ax11, ax12), (ax21, ax22)] = plt.subplots(2, 2, figsize=(10, 6), dpi=144)

fig.suptitle("Four plots, one figure")

ax11.plot(x, y1)
ax11.set_title("1.1: y1")

ax12.plot(x, y2)
ax12.set_title("1.2: y2")

ax21.plot(x, y3)
ax21.set_title("2.1: y3")

ax22.plot(x, y1 - y2 ** 2, color='deeppink')
ax22.set_title("2.2: y1 - y3 ** 2")

fig.tight_layout(pad=3.0)
```

## Line and marker styles

| color   | shortcut | marker                 | shortcut         | line    | shortcut    |
|---------|----------|------------------------|------------------|---------|-------------|
| blude   | b        | point                  | .                | solid   | -           |
| green   | g        | pixel                  | ,                | dashed  | --          |
| red     | r        | circle                 | o                | dashdot | -.          |
| cyan    | c        | triangles              | v ^ < >          | dotted  | :           |
| magenta | m        | unfilled triangles     | 7 6 4 5          | none    | ' ' (space) |
| yellow  | y        | tri                    | '1' '2' '3' '4'  |         |             |
| black   | k        | hexagon (h/v)          | h H              |         |             |
| white   | w        | diamond (regular/thin) | D d              |         |             |
| RGB     | #f0c000  | line (h/v)             | _ &#124;         |         |             |
| grey    | '0.75'   | star                   | *                |         |             |
| named   | teal     | plus                   | +                |         |             |
|         |          | x                      | x                |         |             |
|         |          | square                 | s                |         |             |
|         |          | pentagon               | p                |         |             |

(there are also [named colors](http://matplotlib.org/examples/color/named_colors.html))

```python
xs = np.arange(0, 5, 0.2)
plt.plot(xs, xs, 'g-', marker=7, markersize=10)
plt.plot(xs, xs**2, '|', color='0.4')
# sometime we need to use an explicit ' ' for "no line"
plt.plot(xs, xs**2 + 10, 'r ', marker='1')
plt.plot(xs, xs**3, '*',
         markersize=15,
         markerfacecolor='#f0c000',
         markeredgecolor='r',
         markeredgewidth=2)
```

## Line & Area Charts

https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.plot.html

```python
xs2 = np.arange(0, 10, 0.5)
ys2 = np.exp(-xs2/4) * np.cos(xs2**2)
ys2b = ys2 - 0.2 - np.random.rand(len(xs2)) * 0.5
# there is a condenset syntax for simple cases
#  (eg. line and markers of same color)
# plt.plot(xs2, ys2, 'bv-')
plt.plot(xs2, ys2,
    linestyle='dashed',
    linewidth=0.8,
    color='b',
    marker='v',
    markersize=10,
    markeredgewidth=0.5,
    markeredgecolor='k',
    markerfacecolor='orange',
)

# you COULD plot the actual line for data referenced by
#    fill_between, but you don't have to!
# plt.plot(xs2, ys2b, 'k', linewidth=0.5)

plt.fill_between(xs2, ys2b, ys2, color='g', alpha=0.1)

plt.fill_between(xs2, -1.5, ys2b, color='r', alpha=0.1)
```

### Color by value

```python
from sklearn.datasets import load_iris

iris = load_iris()

plt.figure(figsize=(3, 3))
plt.scatter(
    iris.data[:, 0], iris.data[:, 1],
    c=iris.target,
    edgecolor='k',
    linewidths=0.3,
    cmap=plt.cm.tab20
)
```

### Different markers

```python
from sklearn.datasets import load_iris

iris = load_iris()
targets = np.unique(iris.target)
c1data = iris.data[iris.target == targets[0]]
c2data = iris.data[iris.target == targets[1]]
c3data = iris.data[iris.target == targets[2]]

plt.figure(figsize=(3, 3))
plt.scatter(
    c1data[:, 0], c1data[:, 1],
    c='b',
    marker="x",
    edgecolor='k',
    linewidths=0.3,
    cmap=plt.cm.tab20
)
plt.scatter(
    c2data[:, 0], c2data[:, 1],
    c='brown',
    edgecolor='k',
    linewidths=0.3,
    cmap=plt.cm.tab20
)
plt.scatter(
    c3data[:, 0], c3data[:, 1],
    c='c',
    marker="v",
    edgecolor='k',
    linewidths=0.3,
    cmap=plt.cm.tab20
)
```

### Size by value

```python
from sklearn.datasets import load_iris

iris = load_iris()
targets = np.unique(iris.target)
c3data = iris.data[iris.target == targets[2]]

area = (20 * np.random.rand(len(c3data)))**2

plt.figure(figsize=(3, 3))
plt.scatter(
    c3data[:, 0], c3data[:, 1],
    c='orange',
    s=area,
    marker="o",
    edgecolor='k',
    linewidths=0.3,
    cmap=plt.cm.tab20,
)
```

## Bar charts

### Vertical bars - grouped & stacked

https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.bar.html

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
women_means_b = [10, 5, 20, 36, 12]

hx = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(hx - width/2, men_means,
                width=width, label='Men', color='b')
rects2 = ax.bar(hx + width/2, women_means, bottom=women_means_b,
                width=width, label='Women', color='r')
rects2b = ax.bar(hx + width/2, women_means_b,
                width=width, label='Women B', color='orange')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(hx)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects, bottom_rects=None):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for i, rect in enumerate(rects):
        height = rect.get_height()
        if bottom_rects:
            height += bottom_rects[i].get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, -3.5),  # 3 points vertical offset
                    bbox=dict(facecolor='white'),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2, rects2b)
autolabel(rects2b)

fig.tight_layout()
plt.show()
```

### Horizontal bars - grouped & stacked

https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.barh.html

```python
labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
women_means_b = [10, 5, 20, 36, 12]

hy = np.arange(len(labels))  # the label locations
height = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 8))
rects1 = ax.barh(hy - height/2, men_means,
                height=height, label='Men', color='b')
rects2 = ax.barh(hy + height/2, women_means, left=women_means_b,
                height=height, label='Women', color='r')
rects2b = ax.barh(hy + height/2, women_means_b,
                height=height, label='Women B', color='orange')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_yticks(hy)
ax.set_yticklabels(labels)
ax.legend()

def autolabel(rects, start_rects=None):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for i, rect in enumerate(rects):
        w = rect.get_width()
        if start_rects:
            w += start_rects[i].get_width()
        ax.annotate('{}'.format(w),
                    xy=(w, rect.get_y() + rect.get_height() / 2),
                    xytext=(0, -3.5),  # 3 points vertical offset
                    bbox=dict(facecolor='white'),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2, rects2b)
autolabel(rects2b)

fig.tight_layout()
plt.show()
```

## Pie charts

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

explode = (0, 0.1, 0, 0)  # explode only 2nd part

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, 
       autopct='%.1f%%', startangle=90,
       colors=['r', 'g', 'b', 'orange'],
       labeldistance=1.2,
       textprops={'bbox': dict(facecolor='white', edgecolor='white')},
       wedgeprops=dict(linewidth=2, edgecolor='w'),
)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
vals = np.array([[60., 32.], [37., 40.], [29., 10.]])
print("> vals:"); display(vals)
svals = vals.sum(axis=1)

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(3)*4)  # [0, 4, 8]
inner_colors = cmap(np.array([1, 2, 5, 6, 9, 10]))

ring_size = 0.3

fig, ax = plt.subplots()

ax.pie(svals, labels=svals, radius=1, colors=outer_colors,
       labeldistance=0.78,
       # https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.patches.Wedge.html
       wedgeprops=dict(width=ring_size, edgecolor='w'),
)

ax.pie(vals.flatten(), labels=vals.flatten(),
       labeldistance=0.63,
       radius=1-ring_size, colors=inner_colors,
       wedgeprops=dict(width=ring_size, edgecolor='w'))

plt.show()
```
